```python
l = [ 1,2,3,4,3,2,1,2,5,6]

nl = [] # this is the empty set where am going to insert the values from l and make it original by not adding the same content again and again 
for i in l: #in the value of l making as one by one
    if i not in nl: # if the nl value doesnt have the l's value then 
        nl.append(i) # it will give the output of the unique value and stops if its repeats
print(nl)        
        
    
```

```python
l = [ 1,2,3,4,3,2,1,2,5,6]

f = {} # empty dictionary to store the unique value , the format will be of like { Element : count }

for i in l: # one by one takes the value and make it loop
    if i in f: # checks if the elements already present in the f dict
        f[i] += 1 # if the same number exists then it will add 1 number extra to it 
    else:
        f[i] = 1 #it remains the same 
print(f) # now print the final f dict
```
