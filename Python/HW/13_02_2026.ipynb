```python
num1 = float(input(" enter num1: "))
operator = input("calculation sign : ")
num2 = float(input("enter num2 : "))

if operator == "+":
    ans = num1 + num2
elif operator =="-":
    ans = num1 - num2
elif operator == "/":
    if num2 != 0:
        ans = num1 / num2
    else:
      ans ="its not divisble by 0"
    # exit()
      
elif operator == "*":
    ans = num1 * num2
elif operator == "%":
    if num2 != 0:
        ans = num1 % num2
    else:
        ans ="its not modulus by 0"
    # exit()
        
else:
    ans = "invalid operation"
    
print("result :" ,ans)
    
```
