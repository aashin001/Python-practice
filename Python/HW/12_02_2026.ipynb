```python
skill = input("Enter the skill: ")
experience = input("How many years of experience: ")
location = input("Enter the location: ")


#Infosys
if (skill == "python" and experience != "fresher" and location == "chennai"):
    print("Contratulations! you are eligible for Infosys")
    
#Wipro    
elif (skill == "python" and experience == "fresher" and location == "coimbatore"):
    print("Contratulations! you are eligible for Wipro")
    
#TCS    
elif (skill == "java" and experience == "fresher" and location == "chennai"):
    print(" Contratulations! you are eligible for TCS")

#Cognizant
elif (skill == "java" and experience != "fresher" and location == "bangalore"):
    print(" Contratulations! you are eligible for Cognizant")
else:
    print("sorry you're resume didn't shortlisted")
    
```
